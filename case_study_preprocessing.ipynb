```python
selected_features = mi_X[mi_X['mutual Information'] > 0.01]
selected_features
```

```python
# Updating the 'X' DataFrame with the data whose mutual information > 0.01
X = X[selected_features['Features'].tolist()]
X.head()
```

# Outlier Handling

##Boxplot

```python
plt.figure(figsize = (20,25))

for i, column in enumerate(X.columns):
  plt.subplot(4,4,i+1)
  plt.title(f'Box plot of {column}')
  sns.boxplot(y = X[column])
plt.tight_layout()
plt.show()

# Boxplots before handling outliers
```

## Handling Outliers

```python
def clipping(column):
  Q1 = X[column].quantile(0.25)
  Q3 = X[column].quantile(0.75)
  IQR = Q3 - Q1

  upper_bound = Q1 - (1.5 * IQR)
  lower_bound = Q3 + (1.5 * IQR)

  X[column] = np.clip(X[column], lower_bound, upper_bound)
  return X


# Applying clipping to all columns

for column in X.columns:
  X = clipping(column)
```
